```python
# Cell 8: Parent Class - Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")
```

```python
# Cell 9: Child Class - Student
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)  # Initialize parent attributes
        self.student_id = student_id

    def enroll_course(self, course):
        print(f"Student {self.name} enrolled in {course}")

    # Override display_info
    def display_info(self):
        super().display_info()  # Call parent method
        print(f"Student ID: {self.student_id}")
```

```python
# Cell 10: Child Class - Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)  # Initialize parent attributes
        self.employee_id = employee_id

    def teach_course(self, course):
        print(f"Lecturer {self.name} is teaching {course}")
```

```python
# Cell 11: Demonstration
# Create objects
student = Student("Ama", 20, "ST123")
lecturer = Lecturer("Dr. Mensah", 45, "EMP456")

# Inherited method
student.display_info()
lecturer.display_info()

# Unique methods
student.enroll_course("Python OOP")
lecturer.teach_course("Data Structures")
```
